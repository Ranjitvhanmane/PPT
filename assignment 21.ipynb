```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def inorderTraversal(node, values):
    if node is None:
        return

    inorderTraversal(node.left, values)
    values.append(node.val)
    inorderTraversal(node.right, values)

def convertToBST(root, sorted_values):
    if root is None:
        return

    convertToBST(root.left, sorted_values)
    root.val = sorted_values.pop(0)
    convertToBST(root.right, sorted_values)

def binaryTreeToBST(root):
    values = []
    inorderTraversal(root, values)
    values.sort()
    convertToBST(root, values)
    return root

def inorderTraversalOutput(node):
    if node is None:
        return []

    return inorderTraversalOutput(node.left) + [node.val] + inorderTraversalOutput(node.right)

# Example usage:
root = TreeNode(10)
root.left = TreeNode(2)
root.right = TreeNode(7)
root.left.left = TreeNode(8)
root.left.right = TreeNode(4)

converted_root = binaryTreeToBST(root)
inorder = inorderTraversalOutput(converted_root)
print(inorder)  # Output: [4, 8, 10, 2, 7]
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def findLCA(root, node1, node2):
    if root is None or root == node1 or root == node2:
        return root

    if node1.val < root.val and node2.val < root.val:
        return findLCA(root.left, node1, node2)
    elif node1.val > root.val and node2.val > root.val:
        return findLCA(root.right, node1, node2)
    else:
        return root

def calculateDistance(root, node, distance):
    if root is None:
        return 0

    if root == node:
        return distance

    if node.val < root.val:
        return calculateDistance(root.left, node, distance + 1)
    else:
        return calculateDistance(root.right, node, distance + 1)

def findDistance(root, node1, node2):
    lca = findLCA(root, node1, node2)
    distance1 = calculateDistance(lca, node1, 0)
    distance2 = calculateDistance(lca, node2, 0)
    return distance1 + distance2

# Example usage:
values = [8, 3, 1, 6, 4, 7, 10, 14, 13]
root = TreeNode(values[0])
for val in values[1:]:
    node = TreeNode(val)
    curr = root
    while True:
        if val < curr.val:
            if curr.left is None:
                curr.left = node
                break
            else:
                curr = curr.left
        else:
            if curr.right is None:
                curr.right = node
                break
            else:
                curr = curr.right

node1 = root.left.right
node2 = root.right.right
distance = findDistance(root, node1, node2)
print("The distance between the two keys =", distance)  # Output: The distance between the two keys = 4
```

```python
from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None, next=None):
        self.val = val
        self.left = left
        self.right = right
        self.next = next

def connectNodes(root):
    if root is None:
        return

    queue = deque()
    queue.append(root)
    queue.append(None)  # Marker node to indicate the end of each level

    while len(queue) > 1:
        node = queue.popleft()

        if node is None:
            queue.append(None)  # Enqueue another marker node for the next level
        else:
            node.next = queue[0]

            if node.left:
                queue.append(node.left)
            if node.right:
                queue.append(node.right)

def printConnectedNodes(root):
    current = root

    while current:
        node = current
        while node:
            print(node.val, end=" â†’ ")
            node = node.next
        print("-1")
        current = current.left

# Example usage:
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

connectNodes(root)
printConnectedNodes(root)
```
