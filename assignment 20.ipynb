```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def maxSubtreeSum(root):
    max_sum = float('-inf')  # Initialize max_sum with negative infinity

    def subtreeSum(node):
        nonlocal max_sum

        if node is None:
            return 0

        # Calculate the sum of the left subtree
        left_sum = subtreeSum(node.left)
        # Calculate the sum of the right subtree
        right_sum = subtreeSum(node.right)

        # Calculate the sum of the current subtree
        subtree_sum = node.val + left_sum + right_sum

        # Update max_sum if necessary
        max_sum = max(max_sum, subtree_sum)

        return subtree_sum

    subtreeSum(root)
    return max_sum

# Example usage:
root = TreeNode(1)
root.left = TreeNode(2)
root.right = TreeNode(3)
root.left.left = TreeNode(4)
root.left.right = TreeNode(5)
root.right.left = TreeNode(6)
root.right.right = TreeNode(7)

result = maxSubtreeSum(root)
print(result)  # Output: 28
```

```python
from collections import deque

class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def constructBST(level_order):
    if not level_order:
        return None
    
    queue = deque()
    
    # Create the root node using the first element
    root = TreeNode(level_order[0])
    queue.append(root)
    
    i = 1  # Index to iterate through the level order traversal array
    
    while queue and i < len(level_order):
        node = queue.popleft()
        
        # Create the left child
        left_val = level_order[i]
        if left_val is not None:
            node.left = TreeNode(left_val)
            queue.append(node.left)
        i += 1
        
        if i < len(level_order):
            # Create the right child
            right_val = level_order[i]
            if right_val is not None:
                node.right = TreeNode(right_val)
                queue.append(node.right)
            i += 1
    
    return root

def inorderTraversal(node):
    if node is None:
        return []
    return inorderTraversal(node.left) + [node.val] + inorderTraversal(node.right)

# Example usage:
level_order = [7, 4, 12, 3, 6, 8, 1, 5, 10]
root = constructBST(level_order)

inorder = inorderTraversal(root)
print(inorder)  # Output: [1, 3, 4, 5, 6, 7, 8, 10, 12]
```

```python
class TreeNode:
    def __init__(self, val=0, left=None, right=None):
        self.val = val
        self.left = left
        self.right = right

def canRepresentBST(arr):
    stack = []
    root = TreeNode(float('-inf'))

    for element in arr:
        while stack and element > stack[-1].val:
            root = stack.pop()

        if element < root.val:
            new_node = TreeNode(element)
            root.left = new_node
            stack.append(new_node)
        else:
            new_node = TreeNode(element)
            root.right = new_node
            stack.append(new_node)

    while stack:
        root = stack.pop()

    return "Yes" if not stack else "No"

# Example usage:
arr = [7, 4, 12, 3, 6, 8, 1, 5, 10]
result = canRepresentBST(arr)
print(result)  # Output: Yes
```
